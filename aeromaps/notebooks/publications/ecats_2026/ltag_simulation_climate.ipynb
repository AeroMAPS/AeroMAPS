# Application for ECATS - Reproduction of ICAO/LTAG Scenarios

This document has been published for reproducing the application presented in a conference paper dedicated to AeroMAPS, focuding on the reproduction of ICAO LTAG scenarios. The different assumptions of this prospective scenario are given in the following. 

## Load and perimeter choice

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from aeromaps import create_process
import gc
```

```python
perimeter = "Total"  # "Total" or "International"

if perimeter == "Total":
    configuration_file_path = "data/config_files/total/"
elif perimeter == "International":
    configuration_file_path = "data/config_files/international/"
```

## Process, data and compute

```python
process_is2medium = create_process(
    configuration_file=configuration_file_path + "config_is2medium.yaml"
)
process_is2medium_lowcontrails = create_process(
    configuration_file=configuration_file_path + "config_is2medium_lowcontrails.yaml"
)
process_is2medium_highcontrails = create_process(
    configuration_file=configuration_file_path + "config_is2medium_highcontrails.yaml"
)
process_is2medium_bickel = create_process(
    configuration_file=configuration_file_path + "config_is2medium_bickel.yaml"
)
process_is2medium_alternativefuels = create_process(
    configuration_file=configuration_file_path + "config_is2medium_alternativefuels.yaml"
)
process_is2medium_gwpstar = create_process(
    configuration_file=configuration_file_path + "config_is2medium_gwpstar.yaml"
)
process_is2medium_lwe = create_process(
    configuration_file=configuration_file_path + "config_is2medium_lwe.yaml"
)
process_is2medium_ipcc = create_process(
    configuration_file=configuration_file_path + "config_is2medium_ipcc.yaml"
)
process_is2medium_avoidance = create_process(
    configuration_file=configuration_file_path + "config_is2medium_avoidance.yaml"
)
process_is2medium_avoidance.parameters.operations_contrails_final_gain = 60.0
process_is2medium_avoidance.parameters.operations_contrails_start_year = 2030
process_is2medium_avoidance.parameters.operations_contrails_duration = 10.0
process_is2medium_avoidance.parameters.operations_contrails_final_overconsumption = 2.0
```

```python
# IS2 scenarios
process_is2medium.compute()
process_is2medium_data = process_is2medium.data
process_is2medium.write_json()
del process_is2medium
gc.collect()

process_is2medium_lowcontrails.compute()
process_is2medium_lowcontrails_data = process_is2medium_lowcontrails.data
process_is2medium_lowcontrails.write_json()
del process_is2medium_lowcontrails
gc.collect()

process_is2medium_highcontrails.compute()
process_is2medium_highcontrails_data = process_is2medium_highcontrails.data
process_is2medium_highcontrails.write_json()
del process_is2medium_highcontrails
gc.collect()

process_is2medium_bickel.compute()
process_is2medium_bickel_data = process_is2medium_bickel.data
process_is2medium_bickel.write_json()
del process_is2medium_bickel
gc.collect()

process_is2medium_alternativefuels.compute()
process_is2medium_alternativefuels_data = process_is2medium_alternativefuels.data
process_is2medium_alternativefuels.write_json()
del process_is2medium_alternativefuels
gc.collect()

process_is2medium_gwpstar.compute()
process_is2medium_gwpstar_data = process_is2medium_gwpstar.data
process_is2medium_gwpstar.write_json()
del process_is2medium_gwpstar
gc.collect()

process_is2medium_lwe.compute()
process_is2medium_lwe_data = process_is2medium_lwe.data
process_is2medium_lwe.write_json()
del process_is2medium_lwe
gc.collect()

process_is2medium_ipcc.compute()
process_is2medium_ipcc_data = process_is2medium_ipcc.data
process_is2medium_ipcc.write_json()
del process_is2medium_ipcc
gc.collect()

process_is2medium_avoidance.compute()
process_is2medium_avoidance_data = process_is2medium_avoidance.data
process_is2medium_avoidance.write_json()
del process_is2medium_avoidance
gc.collect()
```

## Results

```python
data_years = np.array((range(1940, 2071)))

fig, ax = plt.subplots()

ax.plot(
    data_years,
    process_is2medium_data["climate_outputs"][["temperature_increase_from_aviation"]].values[
        1940 - 1940 : 2070 - 1940 + 1
    ],
    linestyle="-",
    label="Reference (FaIR)",
)

ax.plot(
    data_years,
    process_is2medium_gwpstar_data["climate_outputs"][
        ["temperature_increase_from_aviation"]
    ].values[1940 - 1940 : 2070 - 1940 + 1],
    linestyle="-",
    label="GWP*",
)

ax.plot(
    data_years,
    process_is2medium_lwe_data["climate_outputs"][["temperature_increase_from_aviation"]].values[
        1940 - 1940 : 2070 - 1940 + 1
    ],
    linestyle="-",
    label="LWE",
)

ax.plot(
    data_years,
    process_is2medium_ipcc_data["climate_outputs"][["temperature_increase_from_aviation"]].values[
        1940 - 1940 : 2070 - 1940 + 1
    ],
    linestyle="-",
    label="IPCC",
)

ax.set(
    xlabel="Year",
    ylabel="Temperature increase (°C)",
)
ax.grid()
plt.xlim(1940, 2070)
plt.title("Model comparison")
plt.legend()
plt.show()
```
