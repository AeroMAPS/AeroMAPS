# How different technology scenarios can impact growth in CO2 emissions ?

Each of the T1-T5 middle scenarios is mapped using the central traffic growth forecast. The T1 scenario shows where CO2
emissions would be with no further improvements in aircraft efficiency and no new technology. This chart does not include
reductions in emissions from the other pillars of action: operations, infrastructure, sustainable fuels or market-based measures.

## Load and run custom processes

First, let's create the process and compute for each of the technology scenarios (T1-T5).

```python
%matplotlib widget
from aeromaps import create_process

import gemseo as gm
from aeromaps.utils.functions import custom_logger_config
custom_logger_config(gm.configure_logger())
```

## Vizualizing single scenarios

Now let's vizualize results for the T1 scenario

```python
process1 = create_process(configuration_file="config_files/config_t1.yaml")
process1.compute()
process1.write_json()
process1.plot("air_transport_co2_emissions")
```

Now for the T2 scenario

```python
process2 = create_process(configuration_file="config_files/config_t2.yaml")
process2.compute()
process2.write_json()
process2.plot("air_transport_co2_emissions")
```

Now for the T3 scenario

```python
process3 = create_process(configuration_file="config_files/config_t3.yaml")
process3.compute()
process3.write_json()
process3.plot("air_transport_co2_emissions")
```

Now for the T4 scenario

```python
process4 = create_process(configuration_file="config_files/config_t4.yaml")
process4.compute()
process4.write_json()
process4.plot("air_transport_co2_emissions")
process4.plot("fuel_shares")
```

Now for the T5 scenario

```python
process5 = create_process(configuration_file="config_files/config_t5.yaml")
process5.compute()
process5.write_json()
process5.plot("air_transport_co2_emissions")
process5.plot("fuel_shares")
```
