```python
import numpy as np
import matplotlib.pyplot as plt

# Define the variables and constants
a = 5.081920045588267e-16
b = 0.09567722144824042
coef_dem = 52282241028707.414

# Define the range of rpk values between 8e12 and 1.5e13
rpk = np.linspace(8e12, 3e13, 500)

# Define the two functions
func1 = a * rpk + b
func2 = coef_dem * rpk ** (1 / -0.9)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(rpk, func1, label="a*rpk + b", color="blue")
plt.plot(rpk, func2, label="coef_dem * rpk^(1/3)", color="green")
for i, el in enumerate(magic_method_view):
    plt.plot(magic_rpk[i], magic_method_view[i], marker="+", markersize=15, label=i)
plt.xlabel("rpk")
plt.ylabel("Values")
plt.title("Comparison of a*rpk+b and coef_dem * rpk^(1/3)")
plt.legend()
plt.grid(True)
plt.show()
```
